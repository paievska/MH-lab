Завдання 3: Створіть масив розміром 6x6, заповніть його випадковими цілими числами від -50 до 50 та знайдіть добуток непарних чисел. 

```python
a = np.random.randint(-50, 51, (6, 6))

prod_odd = np.prod(a[a % 2 != 0])

print("a = ", a)
print("Добуток непарних чисел", prod_odd)
```

Завдання 4: Створіть масив розміром 4x4, заповніть його випадковими дійсними числами в діапазоні від 0 до 1 та знайдіть вектор-рядок, що містить максимальні елементи в кожному стовпці. 

```python
a = np.random.rand(4, 4)

cols_max = a.max(axis = 0)

print("а = ", a)
print("Максимальні елементив кожному стовпці = ", cols_max)
```

Завдання 5: Створіть два одновимірних масиви розміром 7, заповніть їх випадковими цілими числами в діапазоні від -10 до 10 та виконайте на них поелементні операції додавання, віднімання та множення. 

```python
a = np.random.randint(-10, 11, 7)
b = np.random.randint(-10, 11, 7)

print("a = ", a)
print("b = ", b)
print("Додавання елементів матриць = ", a + b)
print("Віднімання елементів матриць = ",a - b)
print("Множення елементів матриць = ",a * b)
```

Завдання 6: Створіть два вектори розміром 7, заповніть їх довільними цілими числами та знайдіть їх скалярний добуток. 

```python
a = np.random.randint(1, 11, size=7)
b = np.random.randint(1, 11, size=7)

dot_a_b = np.dot(a, b)

print("a = ", a)
print("b = ", b)
print("Скалярний добуток векторів дорівнює:", dot_a_b)
```

Завдання 7: Створіть дві матриці розміром 4x2 та 2x2, заповніть їх випадковими цілими числами в діапазоні від 1 до 10 та перемножте їх між собою. 

```python
a = np.random.randint(1, 11, (4, 2))
b = np.random.randint(1, 11, (2, 2))

dot_a_b = np.dot(a, b);

print("a = ", a)
print("b = ", b)
print("Результат перемноження матриць: ", dot_a_b)
```

Завдання 8: Створіть матрицю розміром 2x2, заповніть її випадковими цілими числами в діапазоні від 1 до 7 та знайдіть її обернену матрицю. 

```python
a = np.random.randint(1, 8, size=(2, 2))

print("Вихідна матриця:")
print(a)

# перевірка, чи можна знайти обернену матрицю
if np.linalg.det(a) == 0:
    print("Матриця не має оберненої матриці.")
else:
    # знаходження оберненої матриці
    inv_a = np.linalg.inv(a)
    print("Обернена матриця:")
    print(inv_a)
```

Завдання 9: Створіть матрицю розміром 3x5, заповніть її випадковими цілими числами в діапазоні від 0 до 1 та транспонуйте її. 

```python
a = np.random.randint(0, 2, (3, 5))

print("а = ", a)

a_transp = np.transpose(a)

print("Транспонована матриця:")
print(a_transp)
```

Завдання 10: Створіть матрицю розміром 3x4 та вектор розміром 4, заповніть матрицю випадковими дійсними числами від 0 до 1, а вектор - випадковими цілими числами в діапазоні від 1 до 10, і знайдіть добуток вектора на матрицю. 

```python
a = np.random.rand(3, 4)
b = np.random.randint(1, 11, 4)

dot_a_b = np.dot(a, b)

print("a = ", a)
print("b = ", b)
print("Добуток вектора на матрицю", dot_a_b)
```

Завдання 11: Створіть матрицю розміром 2x3 та вектор розміром 3, заповніть матрицю випадковими дійсними числами від 0 до 1, а вектор - випадковими дійсними числами від 0 до 1, і знайдіть суму елементів матриці, які менше за відповідний елемент вектору. 

```python
a = np.random.rand(2, 3)
b = np.random.rand(3)

res = np.sum(a[a < b])

print("a = ", a)
print("b = ", b)
print("res = ", res)
```

Завдання 12: Створіть дві матриці розміром 8x8, заповніть їх випадковими цілими числами в діапазоні від 1 до 10 та знайдіть різницю елементів першого стовпця першої матриці.

Завдання 13: Створіть дві матриці розміром 7x7, заповніть їх випадковими дійсними числами в діапазоні від -1 до 1 та знайдіть добуток їх елементів. 

```python
a = np.random.rand(7, 7) * 2 - 1
b = np.random.rand(7, 7) * 2 - 1

prod_a = np.prod(a) 
prod_b = np.prod(b)
res = prod_a * prod_b

print("a = ", a)
print("b = ", b)
print("Добуток елементів матриць:", res)
```

Завдання 14: Створіть матрицю розміром 7x7, заповніть її випадковими дійсними числами в діапазоні від -10 до 10 та знайдіть добуток елементів матриці, які більше за їх середнє арифметичне значення. 

```python
a = np.random.rand(7, 7) * 20 - 10

mean_a = np.mean(a)
res = np.prod(a[a > mean_a])

print("a = ", a)
print("Mean value = ", mean_a)
print("Result = ", res)
```

Завдання 15: Створіть матрицю розміром 9x9, заповніть її випадковими дійсними числами в діапазоні від -0.5 до 0.5 та знайдіть вектор-стовпчик, що містить найбільше за модулем ціле значення в кожному рядку матриці. 

```python
a = np.random.rand(9, 9) * 1 - 0.5

rows_max_abs = np.max(np.abs(a), axis = 1)
col_vec = rows_max_abs.reshape((-1, 1))

print("a = ", a)
print("Result: ", col_vec)
```
